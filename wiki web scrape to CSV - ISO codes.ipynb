```python
#get data from Wikipedia page
source = requests.get('https://en.wikipedia.org/wiki/ISO_3166-1_numeric').text #change to your wikipedia link containing list

soup = BeautifulSoup(source, 'lxml')

print(soup.prettify())
```

```python
#start process to extract only the table data
wiki_table = soup.find('table',{'class':'wikitable sortable'})
wiki_table_rows = wiki_table.find_all('tr')
```

```python
#loop through and add elements to data array
data = []
for row in wiki_table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])

df = pd.DataFrame(data, columns=['Code', 'Country_Name', 'Notes']) # Change names for output columns
df = df[~df['Code'].isnull()]  
```

```python
#drop unassigned failsafe
df.drop(df[df['Code']=="Not assigned"].index,axis=0, inplace=True)
```
